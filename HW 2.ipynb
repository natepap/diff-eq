```python
import matplotlib.pyplot as ppl
import numpy as np
import math
```

```python
def eulerMethod(func, timeStep, initialVal, timeLim):
    time = 0
    times = [0]
    vals = [initialVal]
    prevVal = initialVal
    while time < timeLim:
        time += timeStep
        curVal = prevVal + func(prevVal)*timeStep
        prevVal = curVal
        vals.append(curVal)
        times.append(time)
    
    return times, vals
```

```python
x1, y1 = eulerMethod(math.sqrt, 1, 1, 4)
x2, y2 = eulerMethod(math.sqrt, 0.5, 1, 4)
x3, y3 = eulerMethod(math.sqrt, 0.25, 1, 4)
```

```python
ppl.plot(x1,y1, label='Timestep = 1')
ppl.plot(x2,y2, label='Timestep = 0.5')
ppl.plot(x3,y3, label='Timestep = 0.25')

ppl.title('Problem 1')
ppl.xlabel('y')
ppl.ylabel('t')
ppl.legend()
ppl.show()
```

```python
g = 9.81
m = 54
k = 0.18
    
def vel(v):
    return (g-(k/m)*v*v)
```

```python
x, y = eulerMethod(vel, 0.01, 0, 20)
```

```python
ppl.plot(x,y)
ppl.title("Velocity")
ppl.xlabel('t')
ppl.ylabel('v')
ppl.show()
```

```python
y1 = [x/math.sqrt(m*g/k) for x in y]
```

```python
ppl.plot(x,y1)
ppl.plot(x, np.full(len(x), 0.95))
ppl.title("Velocity as Fraction of V_inf")
ppl.xlabel('t')
ppl.ylabel('v')
ppl.show()
```
